```python
import pandas as pd
import json
```

```python
lexic = pd.read_csv("Lexique4.tsv", sep="\t")
```

```python
lexic
```

```python
THRESHOLD = .1
filtered_lexic = lexic.loc[lexic["10_FreqMot"] > THRESHOLD]
```

```python
# reduced_lexic = filtered_lexic.drop_duplicates(subset="3_Phono_IPA", inplace=False)
```

```python
keys = set(filtered_lexic["3_Phono_IPA"])
dict = {k: [] for k in keys}
```

```python
for word in lexic.iterrows():
    try:
        word_data = word[1]
        pronunciation = word_data["3_Phono_IPA"]
        orthography = word_data["1_Mot"]

        dict[pronunciation].append(orthography)
    except KeyError:
        continue
```

```python
dict
```

```python
with open("spelling_dict.json", "w", encoding="utf-8") as f:
    json.dump(dict, f, ensure_ascii=False, indent=2)
```
